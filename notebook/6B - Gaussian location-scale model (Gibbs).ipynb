# The location-scale model

In this example, we consider the very simple task of determining the mean and variance of a dataset.
- We are given $n$ IID observation: $X_1, \dots X_n$
- We are trying to determine their common mean and variance.

In order to do so in a Bayesian setting, we need to choose:
- A conditional model. We will first consider the Gaussian model:
$$ \begin{align} 
        X_i &\sim N( \mu, \sigma^2 ) \\
        X_i &= \mu + \sigma \eta_i 
   \end{align}
   $$
- A prior model for the parameters of the conditional model $\mu,\sigma$. We will consider a Gaussian prior over $\mu$ (with mean $\mu_0$ and standard-deviation $\sigma_0$) and a Gamma prior over $\beta = 1 / \sigma^2$ (with parameters $a_0$ and $b_0$)

The resulting posterior is thus proportional to:

$$ f(\mu,\beta | x_1 \dots x_n) \propto \exp \left( - \frac{1}{2 \sigma_0^2} (\mu - \mu_0)^2  \right) \beta^{a_0-1} \exp( - b_0 \beta) \left[ \prod_{i=1}^{n} \beta^{1/2} \exp \left( - \frac{\beta}{2} (x_i - \mu)^2 \right) \right] $$

which isn't nice in any particular way.

However, this posterior as the following two critical properties:
- The posterior conditional distribution of $\mu | \beta$ is simple: it is a Gaussian !
$$ f(\mu | \beta, x_1 \dots x_n) \propto \exp \left( - \frac{1}{2 \sigma_0^2} (\mu - \mu_0)^2  \right) \exp \left( - \frac{\beta}{2} \left[ \sum_{i=1}^{n} (x_i - \mu)^2 \right] \right) $$
- The posterior conditional distribution of $\beta | \mu$ is also simple: it is a Gamma !
$$ f(\beta | \mu, x_1 \dots x_n) \propto \beta^{a_0-1} \exp( - b_0 \beta) \beta^{n/2} \exp \left( - \frac{\beta}{2}  \left[ \sum_{i=1}^{n} (x_i - \mu)^2 \right] \right) $$

We can thus use the Gibbs Sampling algorithm. This consists in alternatively sampling from these conditional posteriors in order to sample from the posterior.

```python
def mu_conditional_params(beta, mu_0, sigma_0, x_array):
    # this function returns the mean and variance of the Gaussian variable \mu | \beta, x_1 \dots x_n
    n = x_array.shape[0]
    inverse_variance = 1 / sigma_0**2 + n * beta
    variance = 1 / inverse_variance
    mean = ( mu_0 / sigma_0**2 + beta * np.sum( x_array ) ) / inverse_variance # the conditional mean is a weighted average
    return mean, variance

def beta_conditional_params(mu, a_0, b_0, x_array):
    # this function returns the a,b parameters of the Gamma variable \beta | \mu, x_1 \dots x_n
    n = x_array.shape[0]
    sum_of_residuals = np.sum( (x_array - mu)**2 )
    a_param = a_0 + n/2
    b_param = b_0 + sum_of_residuals / 2
    return a_param, b_param

def posterior_gibbs_sampling(x_array,                # data
                   mu_0 = 0, sigma_0 = 10, # prior parameters for \mu
                   a_0 = 1, b_0 = 1,       # prior parameters for \beta = 1 / \sigma^2
                   n_iterations = 10001    # number or Gibbs Sampling iterations
                  ):
    
    # First, we pick the initial values of the parameters. We take them to be the prior means
    try:
        mu = mu_0
        beta = a_0 / b_0
    except: # division by b_0 = 0 might raise an exception in which case we use the standard deviation of the data
        beta = 1 / np.var(x_array)
    
    # Then, we create an array to record the sampled values
    mu_samples   = np.zeros( n_iterations )
    beta_samples = np.zeros( n_iterations )
    
    for i in range(n_iterations):
        # update mu
        conditional_mean, conditional_var = mu_conditional_params(beta, mu_0, sigma_0, x_array)
        mu = conditional_mean + (conditional_var)**0.5 * np.random.randn()
        # update beta
        conditional_a, conditional_b = beta_conditional_params(mu, a_0, b_0, x_array)
        beta = np.random.gamma( conditional_a, 1 / conditional_b )
        
        # save the current value
        mu_samples[i] = mu
        beta_samples[i] = beta
        
    return mu_samples, beta_samples
```

```python
# As a simple example, we can sample from the prior distribution (x_array = [])
mu_samples, beta_samples = posterior_gibbs_sampling( np.array([]) )

plt.subplot(1,2,1)
# The mu samples should have a Gaussian(mu_0, std_0^2) distribution
sb.kdeplot(mu_samples)

plt.subplot(1,2,2)
# The beta samples should have a Beta(a_0, b_0) distribution
sb.kdeplot(beta_samples)

plt.show()
```

## A simple application

Let's use our method on a simple application: the historical data from Michelson trying to measure the speed of light.

One five separate occasions, Michelson measured the speed of light 20 times. The results are reported in the five arrays below. The values correspond to Michelson's measurement, from which 299000 was subtracted for convenience.

```python
michelson = np.zeros( [5,20] )
michelson[0,:] = np.array( [850, 740, 900, 1070, 930, 850, 950, 980, 980, 880, 1000, 980, 930, 650, 760, 810, 1000, 1000, 960, 960] )
michelson[1,:] = np.array( [960, 940, 960, 940, 880, 800, 850, 880, 900, 840, 830, 790, 810, 880, 880, 830, 800, 790, 760, 800] )
michelson[2,:] = np.array( [880, 880, 880, 860, 720, 720, 620, 860, 970, 950, 880, 910, 850, 870, 840, 840, 850, 840, 840, 840] )
michelson[3,:] = np.array( [890, 810, 810, 820, 800, 770, 760, 740, 750, 760, 910, 920, 890, 860, 880, 720, 840, 850, 850, 780] )
michelson[4,:] = np.array( [890, 840, 780, 810, 760, 810, 790, 810, 820, 850, 870, 870, 810, 740, 810, 940, 950, 800, 810, 870] )

print( np.mean(michelson, axis = 1) )

plt.figure()
plt.plot(michelson.flatten(), 'o')

plt.figure()
sb.kdeplot( michelson.flatten() )
plt.show()
```

We will analyze each dataset separately. We still need to choose the prior parameters.
Since, we do not have a good guess for their values, we will choose non-commital values which minimize their influence on the posterior:

$$ \begin{align}
    \sigma_0 &= \infty \\
    a_0      &= 1      \\
    b_0      &= 0
   \end{align} $$


```python
for k in range(5):
    mu_samples, beta_samples = posterior_gibbs_sampling( michelson[k,:], mu_0 = 0, sigma_0 = 10000, a_0 = 1, b_0 = 0)
    
    plt.figure()
    
    plt.subplot(1,2,1)
    sb.kdeplot(mu_samples)
    plt.plot( np.mean(mu_samples), 0, 'rx', ms = 20)
    plt.xlabel("Mu density in experiment " + str(k+1) )
    
    plt.subplot(1,2,2)
    plt.plot(mu_samples[0:200])
    plt.xlabel("First 200 Mu samples")
    
plt.show()
```

```python
# we can also analyze all datasets together    
    
mu_samples, beta_samples = posterior_gibbs_sampling( michelson.flatten(), mu_0 = 0, sigma_0 = 10000, a_0 = 1, b_0 = 0)

plt.figure()

plt.subplot(1,2,1)
sb.kdeplot(mu_samples)
plt.plot( np.mean(mu_samples), 0, 'rx', ms = 20)
plt.xlabel("Mu density (all experiments)" )

plt.subplot(1,2,2)
plt.plot(mu_samples[0:200])
plt.xlabel("First 200 Mu samples")
plt.show()

print("Posterior mean is ", np.mean(mu_samples))
print("Posterior std is ", np.std(mu_samples))
print("Posterior 0.95 credible interval is ", np.mean(mu_samples), " \pm ", 2 * np.std(mu_samples) )
# nb: posterior is Gaussian enough that the Gaussian credible interval is appropriate.
# We could also compute credible interval from the samples with:
# mu_samples.sort()
# left_bound = mu_samples[ np.floor( 0.025 * mu_samples.shape[0] ) ]
# right_bound = mu_samples[ np.ceil( 0.975 * mu_samples.shape[0] ) ]
# print("Posterior 0.95 credible interval is : [", left_bound, ", ", right_bound, "]" )
```

## Conclusions

**On the basis of this analysis with the Gaussian model**, we would conclude that the speed of light is approximately $299 850$ km / s ($\pm 15$)  which is quite removed from the true value of $299 792$.

This is partly due to Michelson's data being biased towards higher values. BUt the discrepancy between the two results is also partly due to the fact that **the Gaussian model is a poor model of this dataset** !

One possibility to see this consists in computing a Q-Q plot.

```python
import statsmodels.api as sm

sm.qqplot( (michelson.flatten() - mu_samples.mean()) / (beta_samples.mean())**(-0.5) )
plt.plot( [-3,3], [-3,3], 'k--')
plt.show()
```

The fit is decent for most values, but there are a few outliers on both sides. These outliers are poorly modeled by a Gaussian model and we will thus need to adjust our model in the future.
