```python
import numpy as np
import matplotlib.pyplot as plt
%load_ext autoreload
%autoreload 2
```

### Generate Dataset 

```python
#Dataset 1: independent
x1 = np.random.normal(0, 5, 100)
y1 = np.random.normal(4, 6, 100)
plt.plot(x1,y1, '*')

#Dataset 2: dependent
x2 = np.random.normal(0,5, 100)
a_true = 3.7
b_true = 1.1
sig_true = 3
y2 = a_true*x2 + np.random.normal(b_true, sig_true, 100)
plt.plot(x2,y2,'*')
```

### Model 1: Location-Scale Model

```python
# Maximum Likelihood Inference
mu = np.mean(y1)
sigma = np.std(y1)
samples1 = np.random.normal(mu,sigma,100)
plt.hist(samples1)
plt.figure()
plt.hist(y1)
```
